**List Methods and Slicing **

**Project Title: Contact Directory **

Scenario:

You have been assigned to build a contact management system where employees can store and retrieve contact details efficiently.

Tasks:

Create a 2D list where each row holds a contact’s name, phone number, and email address.

Allow the user to add a new contact.

Remove the last item Get new items from the users using input() and add those items to the list Sort the list in ascending order based on any values

```python
# Project Title : Contact Directory

# Step 1: Create a 2D List with contact details

contacts = [
    ["mano",9361024847,"mano@gamil.com"],
    ["rajesh",9361034847,"rajesh@gamil.com"],
    ["ranjith",9361044847,"ranjith@gamil.com"]
]
print("Existing Contacts:")

for contact in contacts:
    print(contact)
```

```python
# Allow the user to add a new contact.
name = input("Enter your name: ")
phone = input("Enter phone number: ")
email = input("Enter email address: ")

new_contact = [name,phone,email]

# Add new contact to Contact list
contacts.append(new_contact)

print("\nafter adding new contacs")
for contact in contacts:
  print(contact)
```

```python
# Remove the last item
removed_contact=contacts.pop()

print("\nRemoved contact:",removed_contact)

print('\nConatct after removeing last item :')
for contact in contacts:
  print(contact)
```

```python
#Sort list in ascending order based on name
contacts.sort()

print("\nSorted contact List: ")
for contact in contacts:
  print(contact)
```
