### Case Study– Departmental Data Tracker
--Create two sets: `online_courses` and `offline_courses`.

-- * Perform `union`, `intersection`, `difference`, and `symmetric_difference`.

--4. Create a `frozenset` named `permanent_faculty`.





```python
# Creat two sets

offline_courses = {"Python","Data Science","AI","Cloud Computing"}
online_courses = {"AI","Cyper Security","Networking","Python"}

offline_courses
online_courses
```

```python
# Union
print("Union:",online_courses.union(offline_courses))
```

```python
# Intersection
print("Intersection:",online_courses.intersection(offline_courses))
```

```python
# Difference
print("Online - Offline:",online_courses.difference(offline_courses))
print("Offline - Online:",offline_courses.difference(online_courses))
```

```python
# Symmetric Difference
print("Symmetric Difference:",online_courses.symmetric_difference(offline_courses))
```
