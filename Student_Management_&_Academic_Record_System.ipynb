**Student Management & Academic Record System**

Objective:
Understand how to use Tuples (immutable data) and Lists (mutable, structured data) together
to manage and organize information efficiently.

**Scenario:**
A school wants to maintain student data and marks for analysis.

Each student’s personal details are fixed (Name, Roll Number, Course) → stored in a Tuple.
Each student’s subject marks are modifiable → stored in a List.
A collection of all students’ marks is stored as a 2D List.
For advanced organization, a 3D List is used to represent marks for multiple batches.

```python
# Step 1: Create Tuple for each student (immutable personal info)
student1 = ("Mano",101,"CSE")
student2 = ("Raja",102,"IT")
student3 = ("Kala",103,"EEE")

# Display for Student Details
print("Student Details:")
print(student1)
print(student2)
print(student3)
```

```python
# Step 2: Create 1D List for marks for 3 students
marks1 = [88,47,96]
marks2 = [38,40,90]
marks3 = [28,54,78]

print("\n1D List of Marks")
print(marks1)
print(marks2)
print(marks3)
```

```python
# Step 3: Combine all marks into a 2D List
batch1 =[marks1,marks2,marks3]
print("\n Batch 1 Marks 2D List")
print(batch1)
```

```python
# Step 4: Combine all batches (3D List - two different batches)
# create another 2d called batch2 and store 3 more students marks

batch2 = [
    [18,78,96],
    [89,87,56],
    [63,74,85]
]
print(batch2)

# 3D List containing two batches
all_batches = [batch1, batch2]
print("\n All Batchs 3D List")
print(all_batches)
```
