```python
# Creating a Dictionary
product={'prod_id':101,
         'name':'Laptop',
         'price':35000,
         'stock':25}

print(product)
```

```python
# Accessing Values
product['name']
```

```python
# Adding New Key–Value Pair
product['discount']=1000
product
```

```python
# Modifying a Value
product["price"]=45000
product
```

```python
# keys, values, items
print(product.keys())
print(product.values())
print(product.items())
```

```python
print(product.keys())
```

```python
#get method
product.get("stock")
```

```python
#update
product.update({"price":55000,"stock":67})
product
```

```python
#  pop
product.pop("discount")
```

```python
product.popitem()
product
```

```python
#copy
st=product.copy()
st
product
```

```python
# clear
st.clear()
st
```
